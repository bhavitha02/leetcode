```python
# https://www.educative.io/courses/grokking-coding-interview-in-python/sliding-window-median

from heapq import *


class MedianOfStream:
    def __init__(self):
        self.left = []
        self.right = []

    def balance(self):

        if len(self.left) > len(self.right):
            heappush(self.right, -heappop(self.left))

        if len(self.right) == len(self.left) + 2:
            heappush(self.left, -heappop(self.right))

    # This function should take a number and store it
    def insert_num(self, num):
        if not self.right:
            heappush(self.right, num)
        else:
            m = self.find_median()
            if num >= m:
                heappush(self.right, num)
            else:
                heappush(self.left, -num)

        self.balance()

    def remove(self, num):
        if num >= self.right[0]:
            self.right.remove(num)
            heapify(self.right)
        else:
            self.left.remove(-num)
            heapify(self.left)

        self.balance()

    # This function should return the median of the stored numbers
    def find_median(self):
        if len(self.right) == len(self.left):
            return (self.right[0] - self.left[0]) / 2
        elif len(self.right) > len(self.left):
            return self.right[0]
        else:
            return -self.left[0]


def median_sliding_window(nums, k):
    stream = MedianOfStream()

    for i in range(k):
        stream.insert_num(nums[i])
    out = [stream.find_median()]

    for i in range(k, len(nums)):
        r = nums[i - k]
        stream.remove(r)

        stream.insert_num(nums[i])
        out.append(stream.find_median())

    return out
```

```python
# https://www.educative.io/courses/grokking-coding-interview-in-python/schedule-tasks-on-minimum-machines
import heapq
from heapq import *

# [1,4] , [2,5], [4,6]
# [1,4] , [5, 10], [11, 12]


def tasks(tasks_list):
    tasks_list = sorted(tasks_list, key=lambda x: x[0])
    min_heap = []
    heappush(min_heap, tasks_list[0][1])
    for start, end in tasks_list[1:]:
        if min_heap[0] > start:
            heappush(min_heap, end)
        else:
            heappop(min_heap)
            heappush(min_heap, end)
    return len(min_heap)
```

```python
# https://www.educative.io/courses/grokking-coding-interview-in-python/meeting-rooms-iii

# [[0,5],[1,6],[6,7],[7,8],[8,9]]
import heapq
from heapq import *


def most_booked(meetings, rooms):
    available = [i for i in range(rooms)]
    busy = []
    freq = [0 for _ in range(rooms)]
    meetings.sort(key=lambda x: x[0])
    for start, end in meetings:
        while busy and busy[0][0] <= start:
            end_time, room = heappop(busy)
            heappush(available, room)

        if available:
            room = heappop(available)
            heappush(busy, (end, room))
            freq[room] += 1

        else:
            end_time, room = heappop(busy)
            end_time = end_time + (end - start)
            heappush(busy, (end_time, room))
            freq[room] += 1

    return freq.index(max(freq))
```

```python
# https://www.educative.io/courses/grokking-coding-interview-in-python/find-right-intefor%20rval

from heapq import *


def find_right_interval(intervals):
    # initialise two heaps for start and end times
    start_heap = []
    end_heap = []
    result = []
    for idx, elem in enumerate(intervals):
        heappush(start_heap, (elem[0], idx))
        heappush(end_heap, (elem[1], idx))

    while end_heap:
        end, end_idx = heappop(end_heap)
        # if not start_heap:
        #   result.append((-1,end_idx))
        while start_heap:
            start, start_idx = start_heap[0]
            if start >= end:
                result.append((start_idx, end_idx))
                break
            else:
                heappop(start_heap)

    out = [-1 for _ in range(len(intervals))]
    for start_idx, end_idx in result:
        out[end_idx] = start_idx

    return out
```

```python
# https://www.educative.io/courses/grokking-coding-interview-in-python/largest-number-after-digit-swaps-by-parity


from heapq import *


def largest_integer(num):
    lst = []
    while num > 0:
        digit = num % 10
        lst.append(digit)
        num = num // 10

    lst.reverse()
    even_heap = []
    odd_heap = []
    result = []
    for i in lst:
        if i % 2 == 0:
            heappush(even_heap, -i)
        else:
            heappush(odd_heap, -i)

    for i in lst:
        if i % 2 == 0:
            result.append(-heappop(even_heap))
        else:
            result.append(-heappop(odd_heap))

    n = len(result) - 1
    num = 0
    for i in result:
        num = num + i * 10**n
        n -= 1
    return num
```
